# Collection news  

```python
def news_df(web):
    page=requests.get(web,verify=False)
    soup = bs(page.content)
    quotes = [i.text for i in soup.find_all(class_=['title','tn-main-news-title','news-article__title','news-list__title','tn-hidden'])]
    df = pd.DataFrame()
    df['title']=quotes 
    df['source']=web
    df['date-time']= pd.to_datetime('now')
    return df
```

```python
def skip_elements(elements):
    new_list = [ ]
    i = 1
    for element in elements:
        if i%2!=0:
            c=elements[i]
            new_list.append(c)
        i+=1
    return new_list
```

```python
def add_views(df):
    views= [i.text for i in soup.find_all(class_=['mt-2 text-muted mr-3 text-nowrap'])]
    df['views']=skip_elements(views)
    df['views'] = df['views'].str.replace(r'\D', '')
    return df
```

# Let's Lemmatize our Russian text

```python
m = Mystem()
def lemmatize(text):
    text = text.lower()
    lemm_text = "".join(m.lemmatize(text))
    cleared_text = re.sub(r'[^a-zA-Z]', ' ', lemm_text) 
    return " ".join(cleared_text.split())
```
